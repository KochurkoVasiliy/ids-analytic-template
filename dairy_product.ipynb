```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_products.csv')

df.head()

df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Mass_Net'], bins=20, kde=True)
plt.title('Распределение массы нетто продукта')
plt.xlabel('Масса нетто')
plt.ylabel('Частота')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Корреляционная матрица числовых признаков')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='DairyProduct_PackagingType', data=df, palette='viridis')
plt.title('Распределение типов упаковки продукции')
plt.xlabel('Тип упаковки')
plt.ylabel('Количество')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='DairyProduct_ManufacturerCountry', data=df, palette='Set2')
plt.title('Распределение стран производителей')
plt.xlabel('Страна производителя')
plt.ylabel('Количество')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='DairyProduct_ManufacturerCountry', y='EnergyValueKcal', data=df)
plt.title('Сравнение энергетической ценности по странам производителя')
plt.xlabel('Страна производителя')
plt.ylabel('Энергетическая ценность (ккал)')
plt.xticks(rotation=45)
plt.show()
```
